```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
csv_path = '/kaggle/input/memes/posts.csv'
df = pd.read_csv(csv_path)
print(df.head())
```

```python
print(df.columns)
```

```python
print(df[' over_18'].value_counts())
```

```python
from sklearn.utils import class_weight

y = df[' over_18'].values  # 0 or 1
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y),
    y=y
)
# This returns an array like [weight_for_0, weight_for_1]
class_weights_dict = {i : class_weights[i] for i in range(len(class_weights))}
print("Class Weights:", class_weights_dict)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df['image_filename'] = df['id'].astype(str) + '.png'
```

```python
print(df[['id', 'image_filename']].head())
```

```python
# Ensure column names are stripped
df.columns = df.columns.str.strip()

# Convert the over_18 column to strings
df['over_18'] = df['over_18'].astype(str)

# Verify conversion
print(df['over_18'].dtype)
print(df['over_18'].head())
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=df,
    directory='/kaggle/input/memes/images',
    x_col='image_filename',
    y_col='over_18',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training',
    shuffle=True
)

valid_generator = train_datagen.flow_from_dataframe(
    dataframe=df,
    directory='/kaggle/input/memes/images',
    x_col='image_filename',
    y_col='over_18',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation',
    shuffle=False
)
```

```python
# 1) Load base model without top
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 2) Freeze base model layers initially
base_model.trainable = False

# 3) Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# 4) Compile
model.compile(optimizer=Adam(learning_rate=1e-4), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True, mode='max')
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# Save only the weights in HDF5 format with the correct filename extension
checkpoint = ModelCheckpoint(
    'best_model.weights.h5',
    monitor='val_accuracy',
    save_best_only=True,
    save_weights_only=True,
    mode='max'
)
early_stop = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)

history = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=20,
    callbacks=[checkpoint, early_stop],
    class_weight=class_weights_dict  # if you're using class weights
)
```

```python
val_loss, val_acc = model.evaluate(valid_generator)
print(f"Validation Accuracy: {val_acc:.4f}")
```

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

def preprocess_image(img_path, target_size=(224,224)):
    # Load the image
    img = image.load_img(img_path, target_size=target_size)
    # Convert to array
    img_array = image.img_to_array(img)
    # Normalize to [0, 1]
    img_array = img_array / 255.0
    # Expand dims to match the model's input shape => (1, 224, 224, 3)
    img_array = np.expand_dims(img_array, axis=0)
    return img_array
```

```python
import pandas as pd
import numpy as np

# Let's check the distribution first
print(df['over_18'].value_counts())

# Split into minority (over_18=1) and majority (over_18=0)
df_minority = df[df['over_18'] == '1']
df_majority = df[df['over_18'] == '0']

# Oversample the minority to match the majority
df_minority_oversampled = df_minority.sample(
    n=len(df_majority),
    replace=True,  # allow sampling with replacement
    random_state=42
)

# Combine and shuffle
df_balanced = pd.concat([df_majority, df_minority_oversampled]).sample(frac=1, random_state=42)

print("New distribution after oversampling:")
print(df_balanced['over_18'].value_counts())
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define an ImageDataGenerator with data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 20% for validation
)

# Flow from dataframe for TRAIN
train_generator = train_datagen.flow_from_dataframe(
    dataframe=df_balanced,                  # use the balanced df
    directory='/kaggle/input/memes/images', # adjust if needed
    x_col='image_filename',
    y_col='over_18',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',                    # 'binary' for 0/1 classification
    subset='training',
    shuffle=True
)

# Flow from dataframe for VALIDATION
valid_generator = train_datagen.flow_from_dataframe(
    dataframe=df_balanced,
    directory='/kaggle/input/memes/images',
    x_col='image_filename',
    y_col='over_18',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation',
    shuffle=False
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# 1) Load the base model (pretrained)
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224,224,3))

# 2) Unfreeze the entire base model for fine-tuning
base_model.trainable = True

# 3) Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

# 4) Compile with a low learning rate for fine-tuning
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# Save only the weights in HDF5 format (must end with .weights.h5)
checkpoint = ModelCheckpoint(
    'best_model.weights.h5',
    monitor='val_accuracy',
    save_best_only=True,
    save_weights_only=True,
    mode='max'
)

early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=10,  # can increase if needed
    callbacks=[checkpoint, early_stop]
)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

# Gather predictions on validation data
y_true = []
y_pred = []

valid_generator.reset()  # ensure it starts from the beginning

for i in range(len(valid_generator)):
    X_batch, y_batch = valid_generator[i]
    preds_batch = model.predict(X_batch)
    
    # threshold=0.5
    preds_label = (preds_batch >= 0.5).astype(int).ravel()
    
    y_true.extend(y_batch)
    y_pred.extend(preds_label)

precision = precision_score(y_true, y_pred, pos_label=1)
recall = recall_score(y_true, y_pred, pos_label=1)
f1 = f1_score(y_true, y_pred, pos_label=1)
cm = confusion_matrix(y_true, y_pred)

print("Confusion Matrix:\n", cm)
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1 Score:  {f1:.3f}")
```
